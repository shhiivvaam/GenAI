```python
# Seaborn: Statistical Data Visualization 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
# Numerical Data Plotting

# Rel Plot >> It help understand the relationship between two variables
sns.relplot()
sns.relplot
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df)               # This tells us that: higher the bill heigher the tip
```

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df, hue = 'smoker')    # Using { hue } : We can get the data bifirction that the person is smoker or not
                                                                            # Also we get the insight that: The heighest tip was given by a smoker
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'smoker')          # This is a scatter plot with the desired result and the nature of { relplot() }
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, style = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, style ='smoker', hue = 'time')     # style is the marker
                                                                                        # hue is the color
```

```python
value = np.random.randn(100)
time = np.arange(100)
df = pd.DataFrame(time, value.cumsum())
df
```

```python
time = np.arange(500)
value = np.random.randn(500).cumsum()
df = pd.DataFrame({'time' : time, 'value' : value})
df
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci='sd')     # ci = confidence interval, sd = standard deviation
                                                                                # The light_blue area's width around the plotted blue line, increases to the standard devation
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci='sd', estimator = None)   # estimator : this will remove the spreaded region and will spread the lines accordingly
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'event')
```

```python
from matplotlib.pyplot import errorbar


sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, errorbar='sd', estimator = None, hue = 'event')                      # ci = errorbar : confidence interval
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'event', style = 'region', markers=True, dashes = False)
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = sns.load_dataset('tips'))
```

```python
# Joint Plot :  Draw a plot of two variables with bivariate and univariate graphs

tips = sns.load_dataset('tips')
sns.jointplot(x = tips.total_bill, y = tips.tip)                 # The sctter points, tells about the distribution of the two variables, and the bar plot tells about the remaining distribution data
```

```python
# Regression line - in Joint plot

# Putting the BEST FIT LINE
sns.jointplot(x = tips.total_bill, y = tips.tip, kind = 'reg')          # kind = 'reg'  : regression : it adds the regression line to the graph
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip, kind = 'hex')          # hex : hexagon
```

```python
# Contour Plot          -> This is very much used by the { Metrological Department -> for estimating the center points for: tsunami, earthquake, etc}

sns.jointplot(x = tips.total_bill, y = tips.tip, kind = 'kde')        # kde : kernel density estimator/ estimation
```

```python
# pairplot   >>  it will plot all the pair wise relationship btw dataframes

# It is used to get the knowledge of multiple features
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue ='smoker')
```

```python
sns.pairplot(tips, hue = 'smoker', palette = 'coolwarm')
```

```python
# Box Plot

sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```
