```python
# MERGE AND JOIN

# ----------------MERGE----------------------
# In merge, the merging in two data frames happens on some common column, or we define explicity, in which column we want to merge the two dataframes


# ----------------JOIN----------------------
# In merge, the merging in two data frames happens on the basis of indexes(index values of columns)

df1 = pd.DataFrame(
    {
        'key1': [1, 2, 4, 5, 6],
        'key2': [4, 5, 6, 7, 8],
        'key3': [3, 4, 5, 6, 6]
    }
)

df1
```

```python
df2 = pd.DataFrame(
    {
        'key1' : [1, 2, 45, 6, 67],
        'key4' : [56, 5, 6, 7, 8],
        'key5' : [3, 56, 5, 6, 6]
    }
)
df2
```

```python
pd.merge(df1 ,df2, how='inner')        # This works as, it identifies whuch all the columns are same(common) and then it check for the common(same) values in the column to perform the merge
# pd.merge(df1 ,df2, how='left')
# pd.merge(df1 ,df2, how='right')
```

```python
# pd.merge(df1 ,df2, how='outer')   # all the keys from both tha data frames will be present
pd.merge(df1 ,df2, how='cross')   # every row of first dataframe will  be there with each row of the second frame
```

```python
# To merge the dataframes with respect to certain column of both the DataFramesp
pd.merge(df1, df2, how='left', left_on="key2", right_on="key4")
```

```python
# JOIN

df1 = pd.DataFrame(
    {
        'key1' : [1, 2, 4, 5, 6],
        'key4' : [4, 5, 6, 7, 8],
        'key5' : [3, 4, 5, 6, 6]
    },
    index = ['a', 'b', 'c', 'd', 'e']
)
df1
```

```python
df2 = pd.DataFrame(
    {
        'key1' : [1, 2, 45, 6, 67],
        'key4' : [56, 5, 6, 7, 8],
        'key5' : [3, 56, 5, 6, 6]
    },
    index = ['a', 'b', 'h', 'i', 'j']
)
df2
```

```python
# df1.join(df2, how = "inner")   # Join columns of another DataFrame (with regards to some common(attributes -> columns))

# df1.join(df2, on='common_column', how='inner')
df1.join(df2, how="inner", lsuffix='_df1', rsuffix='_df2')
```
