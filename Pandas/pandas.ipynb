```python
# Pandas -> It is a fast, powerfull, flexible and easy to use open source data analysis and manipulation tool built on top of python.
import pandas as pd
```

```python
df = pd.read_csv('services.csv')   # used to read { comma seperated values -> CSV file} into a DataFrame
df.head(2)
```

```python
# header -> when we want to include the header(title section) of the csv file as a row

df = pd.read_csv("services.csv", header=None)    # the top section will now be treated as the row only and not the title(header)
df.head(3)

# internel integer indexes are provided by pandas
# and this not only generates the indexes for columns, but for rows also
```

```python
# skiprows       -> when we want to skip some starting rows

df = pd.read_csv("services.csv", skiprows=1)    # this will result in skipping the first { 3 } rows including the header section(title section -> top level section in a CSV file where the title for the rows are defined)
df.head(3)
```

```python
# when we want to read only some desired columns

df = pd.read_csv("services.csv", usecols=['program_id', 'application_process'])    
df.head(5)
```
