```python
# Line plot

# Whenever time components is involved on x axis
# depict the relation between two variables in the trend

from matplotlib.lines import lineStyles


x = [1, 2, 3, 4, 5]
y = [5, 1, 7, 8, 2]
# plt.plot(x, y)
plt.plot(x, y, color = "red", marker ="o", linestyle = '--', linewidth = 3, markersize = 9)

plt.title('Line Plot to understand the syntax')
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker ="o", linestyle = '--', linewidth = 3, markersize = 9)

plt.title('Line Plot to understand the syntax')
plt.xlabel("x points")
plt.ylabel("y points")
plt.grid()
plt.show()
```

```python
from cProfile import label


x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 2]


plt.plot(x, y1, color = "red", marker ="o", linestyle = '--', linewidth = 3, markersize = 9, label="line y1")
plt.plot(x, y2, color = "b", marker ="o", linestyle = '--', linewidth = 3, markersize = 10, label="line y2")

plt.title('Multiples Lines in One Plot')
plt.xlabel("x Data Axis")
plt.ylabel("y Data Axis")

plt.legend()
plt.show()
```

```python
# stacked histogram

data1 = np.random.normal(0, 1, 5000)
data2 = np.random.normal(0, 1, 5000)
```

```python
print(data1)
print(data2)
```

```python
# Stacked Histogram   -> Adding Multiple Histogram together

plt.hist([data1, data2], bins=20, color=['r', 'b'], label=['dist1', 'dist2'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')
```

```python
# pie chart  > To understand the percent distribution of data
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]
plt.pie(students, labels = langs)
```

```python
plt.pie(students, labels = langs, autopct='%1.1f%%')         # autopct : to show the percentage of the data in the pie chart
```

```python
# explode
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]
plt.pie(students, labels = langs, autopct='%1.1f%%', explode = (0.1, 0.2, 0.3))         # autopct : to show the percentage of the data in the pie chart
```

```python
# explode
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]
# explode = (0.1, 0.2, 0.3)
plt.pie(students, labels = langs, autopct='%1.1f%%', explode = (0.1, 0.2, 0.3), shadow = True)
```
